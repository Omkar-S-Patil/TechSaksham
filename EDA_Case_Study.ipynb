```python
def plot_histograms(df, numerical_cols):
    for col in numerical_cols:
        plt.figure(figsize=(9, 4))
        sns.histplot(df[col], bins=20, kde=True, color="blue")
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()
numerical_cols = ['rate', 'votes', 'approx_cost(for two people)']
plot_histograms(df, numerical_cols)
```

```python
def plot_boxplots(df, numerical_cols):
    plt.figure(figsize=(12, 10))
    for col in numerical_cols:
        plt.figure(figsize=(5, 4))
        sns.boxplot(y=df[col], color="cyan")
        plt.title(f'Boxplot of {col}')
        plt.ylabel(col)
        plt.show()
numerical_cols = ['rate', 'votes', 'approx_cost(for two people)']
plot_boxplots(df, numerical_cols)
```

```python
def plot_bar_charts(df, categorical_cols):
    for col in categorical_cols:
        plt.figure(figsize=(8,5))
        sns.countplot(x=df[col], palette="coolwarm")
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=30)
        plt.show()
categorical_cols = ['online_order', 'book_table',  'listed_in(type)']
plot_bar_charts(df, categorical_cols)
```

```python
def plot_pie_chart(df, column):
    data = df[column].value_counts()  # Get category counts
    labels = data.index  # Get category names
    plt.figure(figsize=(6, 6))
    plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'orange'])
    plt.title(f'Distribution of {column}')
    plt.show()
plot_pie_chart(df, 'listed_in(type)')
```

```python
df.drop(labels=['name', 'listed_in(type)'],axis=1,inplace=True)
```

```python
df.drop(labels=['book_table'],axis=1,inplace=True)
```

```python
df['online_order'] = df['online_order'].map({'Yes': 1, 'No': 0})
```

```python
df['rate'] = df['rate'].astype(str)
```

```python
df['rate'] = df['rate'].apply(lambda x: x.split('/')[0] if isinstance(x, str) else '0').astype(float)
```

```python
columns_to_plot = ['online_order', 'rate', 'votes', 'approx_cost(for two people)']
numeric_columns = df[columns_to_plot]
corr_matrix = numeric_columns.corr()
```

```python
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
